```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("data.csv") #System2(depth=2)
df2 = pd.read_csv("data1.csv") #System2(depth=1)
df3 = pd.read_csv("data2.csv")#System1
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 10, 10
```

```python
mean1 = df1["time"].mean()
mean3 = df3["time"].mean()
```

```python
mean2 = df2["time"].mean()
```

```python
plt.plot(df1["time"], label='System2(depth=2)')
plt.plot(df3["time"]  ,label='System2(depth=1)')
plt.plot(df2["time"]  ,label='System1')
plt.axhline(y=mean1, color='r', linestyle='-')
plt.axhline(y=mean3, color='orange', linestyle='-')
plt.axhline(y=mean2, color='g', linestyle='-')
plt.xlabel('Game Number')
plt.ylabel('Time')
plt.plot()
plt.legend()
```
