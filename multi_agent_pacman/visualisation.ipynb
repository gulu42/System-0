```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data.csv")
```

```python
df.head()
```

```python
mean = df["time"].mean()
```

```python
plt.plot(df["time"])
plt.axhline(y=mean, color='r', linestyle='-')
plt.xlabel('Game Number')
plt.ylabel('Time')
plt.plot()
```

```python
mean = df["score"].mean()
plt.plot(df["score"])
plt.axhline(y=mean, color='g', linestyle='-')
plt.axhline(y=0, color='r', linestyle='-')
plt.xlabel('Game Number')
plt.ylabel('Score')
plt.plot()
```
