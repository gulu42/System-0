```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("data.csv") #System2(depth=2)
df2 = pd.read_csv("data1.csv") #System2(depth=1)
df3 = pd.read_csv("data2.csv")#System1
df4 = pd.read_csv("data_sys0.csv")#System 0 random
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 10, 10
```

```python
colours = {"sys2_d2":'r',"sys2_d1":'orange',"sys1":'b',"sys0_rand":'g'}
```

```python
mean1 = df1["time"].mean()
mean2 = df2["time"].mean()
mean3 = df3["time"].mean()
mean4 = df4["time"].mean()

plt.plot(df1["time"], label='System2(depth=2)', color=colours["sys2_d2"])
plt.axhline(y=mean1, color=colours["sys2_d2"], linestyle='--')

plt.plot(df2["time"]  ,label='System2(depth=1)',color=colours["sys2_d1"])
plt.axhline(y=mean2, color=colours["sys2_d1"], linestyle='--')

plt.plot(df3["time"]  ,label='System1',color=colours["sys1"])
plt.axhline(y=mean3, color=colours["sys1"], linestyle='--')

plt.xlabel('Game Number')
plt.ylabel('Time')
plt.plot()
plt.legend()
```

```python
mean1 = df1["score"].mean()
mean2 = df2["score"].mean()
mean3 = df3["score"].mean()
mean4 = df4["score"].mean()
print(mean1,mean2,mean3,mean4)
```

```python
mean1 = df1["score"].mean()
mean2 = df2["score"].mean()
mean3 = df3["score"].mean()
mean4 = df4["score"].mean()

plt.plot(df1["score"],label='System2(depth=2)',color=colours["sys2_d2"])
plt.axhline(y=mean1, color=colours["sys2_d2"], linestyle='--')

# plt.plot(df2["score"]  ,label='System2(depth=1)',color=colours["sys2_d1"])
# plt.axhline(y=mean2, color=colours["sys2_d1"], linestyle='--')

plt.plot(df3["score"]  ,label='System1', color=colours["sys1"])
plt.axhline(y=mean3, color=colours["sys1"], linestyle='--')

plt.axhline(y=0, color='black', linestyle='-')
plt.xlabel('Game Number')
plt.ylabel('Score')
plt.plot()
plt.legend()
```

```python
mean1 = df1["score"].mean()
mean2 = df2["score"].mean()
mean3 = df3["score"].mean()
mean4 = df4["score"].mean()

plt.plot(df1["score"],label='System2(depth=2)',color=colours["sys2_d2"])
plt.axhline(y=mean1, color=colours["sys2_d2"], linestyle='--')

# plt.plot(df2["score"]  ,label='System2(depth=1)',color=colours["sys2_d1"])
# plt.axhline(y=mean2, color=colours["sys2_d1"], linestyle='--')

plt.plot(df3["score"]  ,label='System1', color=colours["sys1"])
plt.axhline(y=mean3, color=colours["sys1"], linestyle='--')

plt.plot(df4["score"]  ,label='System0', color=colours["sys0_rand"])
plt.axhline(y=mean4, color=colours["sys0_rand"], linestyle='--')

plt.axhline(y=0, color='black', linestyle='-')
plt.xlabel('Game Number')
plt.ylabel('Score')
plt.plot()
plt.legend()
```
